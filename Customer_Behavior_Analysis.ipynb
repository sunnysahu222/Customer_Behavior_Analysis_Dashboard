```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("customer_behavior.db")  # creates a file
```

```python
table_name = "customer_data"

df.to_sql(table_name, conn, if_exists='replace', index=False)

print(f"Data loaded into table '{table_name}' successfully.")
```

```python
query_result = pd.read_sql("SELECT * FROM customer_data LIMIT 20;", conn)
query_result
```

```python
# what is total revenue generated by male vs female cutomer
query_result = pd.read_sql("SELECT gender, SUM(purchase_amount) AS total_revenue FROM customer_data GROUP BY gender;", conn)
query_result
```

```python
# which customer used a discount but still spend more than the spent amount
query_result = pd.read_sql("select customer_id,purchase_amount from customer_data where discount_applied= 'Yes' and purchase_amount >=(select avg(purchase_amount ) from customer_data); ",conn)
query_result
```

```python
# which are top  5 prducts with highest average rating
query_result = pd.read_sql("SELECT item_purchased, AVG(review_rating) AS avg_rating FROM customer_data GROUP BY item_purchased ORDER BY avg_rating DESC LIMIT 5;", conn)
query_result
```

```python
# compare average purchase amount between  standard and express shipping
query_result = pd.read_sql("SELECT shipping_type, round(AVG(purchase_amount),2) AS avg_purchase_amount FROM customer_data where shipping_type in('Express','Standard') GROUP BY shipping_type;", conn)
query_result
```

```python
# do subsried customer spend more ? compare average spend and total evenue between subscribrs and non-subscribers
query_result = pd.read_sql("SELECT subscription_status, round(AVG(purchase_amount),2) AS avg_purchase_amount, SUM(purchase_amount) AS total_revenue FROM customer_data GROUP BY subscription_status;", conn)
query_result
```

```python
# which 5 product have highest percentage of purchase with discount applied
query_result = pd.read_sql("select item_purchased,round(100 *sum(case when discount_applied='Yes' THEN 1 ELSE 0 END)/COUNT(*),2) AS discount_rate from customer_data group by item_purchased order by discount_rate desc limit 5;",conn)
query_result
```

```python
# Segment customer into New ,Returning and loyal basd on total num of purchase and show the count of each segment
query = """
WITH customer_type AS (
    SELECT
        customer_id,
        previous_purchases,
        CASE
            WHEN previous_purchases = 1 THEN 'New'
            WHEN previous_purchases BETWEEN 2 AND 10 THEN 'Returning'
            ELSE 'Loyal'
        END AS customer_segment
    FROM customer_data
)
SELECT
    customer_segment,
    COUNT(*) AS segment_count
FROM customer_type
GROUP BY customer_segment;
"""
query_result = pd.read_sql(query, conn)
query_result
```

```python
# what is top 3 most purchased products within each category
query = """ with item_count as (select category,item_purchased,count(customer_id)
as  total_orders,row_number() over(partition by category order by count(customer_id) desc) as item_rank from customer_data group by category,item_purchased)

select item_rank,category,item_purchased,total_orders from item_count where item_rank <=3;"""
query_result = pd.read_sql(query, conn)
query_result
```

```python
# Are customers who are repeat buyers (more than 5 previous purchases) also likely to subscribe?
query = """
select subscription_status ,count(customer_id) as repeat_buyers from customer_data
             where previous_purchases >5 group by subscription_status;
"""
query_result = pd.read_sql(query, conn)
query_result
```

```python
# what is the revenue contribution of each age
query = """ select age_group,sum(purchase_amount) as total_revenue from customer_data group by age_group order by total_revenue desc;"""
query_result = pd.read_sql(query, conn)
query_result
```
